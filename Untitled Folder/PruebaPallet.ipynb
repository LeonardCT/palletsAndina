```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage import filters
from skimage import feature
from scipy.signal import convolve2d
from scipy.ndimage import median_filter
from matplotlib import pyplot as plt
from skimage.color import rgb2gray

from skimage.morphology import convex_hull_image
```

```python
img = imread('P_A8.jpeg')
```

```python
plt.imshow(img)
plt.show()
```

```python
# Convierte la imagen a YCrCb
ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

# Extrae el canal Y (luminancia)
y_channel = ycrcb[:, :, 0]

# Aplica umbralizacin para separar el fondo del objeto
_, thresholded = cv2.threshold(y_channel, 180, 250, cv2.THRESH_BINARY)

# Encuentra el contorno del objeto despus de eliminar el fondo
contours, _ = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibuja los contornos en una imagen en blanco
result = np.zeros_like(img)

cv2.drawContours(result, contours, -1, (0, 255, 0), 2)  # Dibuja los contornos en verde

# Mostrar la imagen con el fondo eliminado usando matplotlib
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Fondo Eliminado")
plt.axis('off')
plt.show()
```

```python
# Define las coordenadas de la esquina superior izquierda y la esquina inferior derecha del rea que deseas recortar
x1, y1 = 75, 100  # Coordenadas de la esquina superior izquierda
x2, y2 = 1450, 1000  # Coordenadas de la esquina inferior derecha

# Recorta la imagen
recortada = img[y1:y2, x1:x2]

# Muestra la imagen recortada
plt.imshow(recortada)
plt.show()
```

```python
# Convierte la imagen a YCrCb
ycrcb = cv2.cvtColor(recortada, cv2.COLOR_BGR2YCrCb)

# Extrae el canal Y (luminancia)
y_channel = ycrcb[:, :, 0]

# Aplica umbralizacin para separar el fondo del objeto
_, thresholded = cv2.threshold(y_channel, 170, 200, cv2.THRESH_BINARY)

# Encuentra el contorno del objeto despus de eliminar el fondo
contours, _ = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibuja los contornos en una imagen en blanco
result = np.zeros_like(recortada)

cv2.drawContours(result, contours, -1, (0, 255, 0), 2)  # Dibuja los contornos en verde

# Mostrar la imagen con el fondo eliminado usando matplotlib
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Fondo Eliminado")
plt.axis('off')
plt.show()
```

```python
# Convierte la imagen binarizada a formato booleano
binary_image = thresholded.astype(bool)

# Aplica el convex hull a la imagen binarizada
hull = convex_hull_image(binary_image)

# Mostrar la imagen resultante despus de aplicar el convex hull
plt.imshow(hull, cmap='gray')
plt.title("Convex Hull")
plt.axis('off')
plt.show()
```

```python
canny_edge = cv2.Canny((hull * 255).astype(np.uint8), 100, 200)  # Ajusta los umbrales segn sea necesario

# Muestra la imagen del borde obtenido con Canny
plt.imshow(canny_edge, cmap='gray')
plt.title("Borde del Convex Hull con Canny")
plt.axis('off')
plt.show()
```
