```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage import filters
from skimage import feature
from scipy.signal import convolve2d
from scipy.ndimage import median_filter
from matplotlib import pyplot as plt
from skimage.color import rgb2gray

from skimage.morphology import convex_hull_image
```

```python
img = imread('P_A8.jpeg')
```

```python
plt.imshow(img)
plt.show()
```

```python
# Convierte la imagen a YCrCb
ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

# Extrae el canal Y (luminancia)
y_channel = ycrcb[:, :, 0]

# Aplica umbralización para separar el fondo del objeto
_, thresholded = cv2.threshold(y_channel, 180, 250, cv2.THRESH_BINARY)

# Encuentra el contorno del objeto después de eliminar el fondo
contours, _ = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibuja los contornos en una imagen en blanco
result = np.zeros_like(img)

cv2.drawContours(result, contours, -1, (0, 255, 0), 2)  # Dibuja los contornos en verde

# Mostrar la imagen con el fondo eliminado usando matplotlib
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Fondo Eliminado")
plt.axis('off')
plt.show()
```

```python
# Define las coordenadas de la esquina superior izquierda y la esquina inferior derecha del área que deseas recortar
x1, y1 = 75, 100  # Coordenadas de la esquina superior izquierda
x2, y2 = 1450, 1000  # Coordenadas de la esquina inferior derecha

# Recorta la imagen
recortada = img[y1:y2, x1:x2]

# Muestra la imagen recortada
plt.imshow(recortada)
plt.show()
```

```python
# Convierte la imagen a YCrCb
ycrcb = cv2.cvtColor(recortada, cv2.COLOR_BGR2YCrCb)

# Extrae el canal Y (luminancia)
y_channel = ycrcb[:, :, 0]

# Aplica umbralización para separar el fondo del objeto
_, thresholded = cv2.threshold(y_channel, 170, 200, cv2.THRESH_BINARY)

# Encuentra el contorno del objeto después de eliminar el fondo
contours, _ = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibuja los contornos en una imagen en blanco
result = np.zeros_like(recortada)

cv2.drawContours(result, contours, -1, (0, 255, 0), 2)  # Dibuja los contornos en verde

# Mostrar la imagen con el fondo eliminado usando matplotlib
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Fondo Eliminado")
plt.axis('off')
plt.show()
```

```python
# Convierte la imagen binarizada a formato booleano
binary_image = thresholded.astype(bool)

# Aplica el convex hull a la imagen binarizada
hull = convex_hull_image(binary_image)

# Mostrar la imagen resultante después de aplicar el convex hull
plt.imshow(hull, cmap='gray')
plt.title("Convex Hull")
plt.axis('off')
plt.show()
```

```python
canny_edge = cv2.Canny((hull * 255).astype(np.uint8), 100, 200)  # Ajusta los umbrales según sea necesario

# Muestra la imagen del borde obtenido con Canny
plt.imshow(canny_edge, cmap='gray')
plt.title("Borde del Convex Hull con Canny")
plt.axis('off')
plt.show()
```

```python
recortada_test=recortada.copy()
canny_test=canny_edge.copy()
plt.imshow(recortada_test)
```

```python

corners=cv2.goodFeaturesToTrack(canny_test,4,0.1,400)
corners=np.intp(corners)

esquinas_guardadas=[]


for corner in corners:
    x,y=corner.ravel()
    cv2.circle(recortada_test,(x,y),5,(255,0,0),-1)
    esquinas_guardadas.append((x, y))
    
esquinas_guardadas = np.array(esquinas_guardadas)
print("Coordenadas de las esquinas guardadas:")
print(esquinas_guardadas)
    
plt.imshow(recortada_test)
```

```python
corner_coordinates = np.array([[1345, 809], [12, 833], [1131, 59], [242, 56]], dtype=np.float32)

# Ordena las coordenadas de acuerdo con tu lógica
sorted_corner_coordinates = np.zeros_like(corner_coordinates)

# Encuentra la esquina superior izquierda (con los valores mínimos en x e y)
sorted_corner_coordinates[0] = corner_coordinates[np.argmin(np.sum(corner_coordinates, axis=1))]

# Encuentra la esquina superior derecha (con el valor mínimo en y y el valor máximo en x)
sorted_corner_coordinates[1] = corner_coordinates[np.argmin(corner_coordinates[:, 1])]
sorted_corner_coordinates[1, 0] = np.max(corner_coordinates[:, 0])

# Encuentra la esquina inferior derecha (con los valores máximos en x e y)
sorted_corner_coordinates[2] = corner_coordinates[np.argmax(np.sum(corner_coordinates, axis=1))]

# Encuentra la esquina inferior izquierda (con el valor máximo en y y el valor mínimo en x)
sorted_corner_coordinates[3] = corner_coordinates[np.argmax(corner_coordinates[:, 1])]
sorted_corner_coordinates[3, 0] = np.min(corner_coordinates[:, 0])
# Define las coordenadas de destino para el warping
# Por ejemplo, puedes definir un rectángulo con un tamaño fijo para la imagen de salida
# Aquí, estamos usando un rectángulo de 400x300 píxeles
width, height = 400, 300
destination_coordinates = np.array([[0, 0], [width - 1, 0], [width - 1, height - 1], [0, height - 1]], dtype=np.float32)

# Calcula la matriz de transformación
transformation_matrix = cv2.getPerspectiveTransform(sorted_corner_coordinates, destination_coordinates)

# Lee la imagen original

# Aplica la transformación de perspectiva a la imagen
image_warped = cv2.warpPerspective(recortada_test, transformation_matrix, (width, height))

plt.imshow(recortada_test, cmap='gray')  # Mostrar la imagen original
plt.title('Imagen Original')
plt.show()

plt.imshow(image_warped, cmap='gray')  # Mostrar la imagen transformada
plt.title('Imagen Transformada')
plt.show()
```
